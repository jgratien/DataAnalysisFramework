```python
import sys
sys.path.append("../TimeSeriesTools")
import utils
import time
from kafka import KafkaProducer
```

```python
def find_all_data(collection,volume):
    data_list=[]
    i=0
    for d in collection.find():
        data_list.append(d)
        i+=1
        if(i==volume):
            print(len(data_list), " documents found")
            return data_list       
```

```python
def convert_data(data):
    msg_1 = list(data['data']['Elements'].values())
    msg_2 = list(data['data']['Naneos'].values())
    time_1 = str(data['updatedAt'])
    time_2 = str(data['createdAt'])
    msg_list = msg_1 + msg_2 
    conv = lambda i : str(i) or ''
    res = [conv(i) for i in msg_list] 
    msg = str(res[0])
    for i in range(1,len(res)):
        #print(str(res[i]))
        msg = msg + ";" + str(res[i])
    msg = msg + ";" + time_1 + ";" + time_2
    return msg
```

```python
def producer_pollution(data,topic):
    producer = KafkaProducer(bootstrap_servers="localhost:9092")
    for i in range(0, len(data)):
        msg = convert_data(data[i])
        msg=str(i)+":"+msg
        print(msg)
        producer.send(topic,msg.encode('utf8'))
        time.sleep(0.1)
```
